```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as special
from scipy.optimize import curve_fit
import seaborn as sns
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
dfM = pd.read_csv('Datos_limpios_Mexico_V3.csv')
```

# Convertir a dicotomica

```python
Vars_Indep=dfM[['bedrooms','beds','room_type']]
Var_Dep=dfM['accommodates'] #variable objetivo
```

```python
X=Vars_Indep
y=Var_Dep
```

¿Qué valor usar en test_size?
En datasets grandes, mantener un 0.2 o incluso 0.1.
En datasets pequeños, mantener un 0.3 o más.
¿Qué valor usar en random_state?
Usualmente se recomienda mantener un 42 para no variar la division de los datos y usarlos varias veces

```python
#manejar el desbalance con stratify=y
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=40,stratify=y)
```

```python
#cuantos valores tiene cada repsuesta de host is superhost
#util para notar si hay desbalanceo en los datos
conteo = dfM['host_response_rate'].value_counts()
conteo
```

```python
escalar=StandardScaler()
```

```python
X_train=escalar.fit_transform(X_train)
X_test=escalar.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
algoritmo=LogisticRegression()
```

```python
algoritmo.fit(X_train,y_train)
```

```python
y_pred=algoritmo.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
matriz=confusion_matrix(y_test,y_pred)
print('Matriz de confusion')
print(matriz)
#30
```

```python
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred , average='binary',pos_label=1)
print('Precision del modelo')
print(precision)
```

```python
#calcular exactitu del modelo
from sklearn.metrics import accuracy_score

exactitud=accuracy_score(y_test,y_pred)
print('Exactitud del modelo')
print(exactitud)
```

Difrencia netre exactitud y precisión 
Exactitud te da una medida general de la eficacia del modelo en todas las clases.

Precisión te da una medida de la eficacia del modelo solo en la clase positiva especificada, ignorando los resultados de la clase negativa excepto en la medida en que contribuyen a los falsos positivos.

```python
#calcular sensibilidad del modelo
from sklearn.metrics import recall_score
sensibilidad=recall_score(y_test,y_pred,average='binary',pos_label=0)
print('Sensibilidad del modelo')
print(sensibilidad)
```
