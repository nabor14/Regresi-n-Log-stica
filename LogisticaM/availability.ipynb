```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as special
from scipy.optimize import curve_fit
import seaborn as sns
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
dfM=pd.read_csv('Datos_limpios_Mexico.csv')
dfM= dfM.drop(['Unnamed: 0'], axis=1)
dfM.head(5)
```

```python
unico=np.unique(dfM['availability_30'])
unico
```

```python
unico=np.unique(dfM['instant_bookable'])
unico
```

```python
unico=np.unique(dfM['host_listings_count'])
unico
```

```python
dfM['instant_bookable'] = dfM['instant_bookable'].map({'t': 1, 'f': 0})
```

```python
Vars_Indep=dfM[['availability_30','instant_bookable','host_listings_count']]
Var_Dep=dfM['has_availability'] #variable objetivo
```

```python
X=Vars_Indep
y=Var_Dep
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.3, random_state=None)
```

```python
#se escalan todos los datos
escalar= StandardScaler()
```

```python
#para realizar el escalamiento de las variables X tanto de entrenamiento como de prueba 
X_train= escalar.fit_transform(X_train)
X_test= escalar.transform(X_test)
```

```python
#definimos el algoritmo a utilizar
from sklearn.linear_model import LogisticRegression
algoritmo= LogisticRegression()
```

```python
algoritmo.fit(X_train,y_train)
```

```python
y_pred=algoritmo.predict(X_test)
y_pred
```

```python
modelo = LogisticRegression(class_weight='balanced')
modelo.fit(X_train, y_train)
y_pred = modelo.predict(X_test)
```

```python
#matriz de confusion
from sklearn.metrics import confusion_matrix
matriz= confusion_matrix(y_test, y_pred)
print('Matriz de Confusion:')
print(matriz)
```

```python
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred , average='binary',pos_label='t')
print('Precision del modelo')
print(precision)
```

```python
#calcular exactitu del modelo
from sklearn.metrics import accuracy_score

exactitud=accuracy_score(y_test,y_pred)
print('Exactitud del modelo')
print(exactitud)
```

```python
#calcular sensibilidad del modelo
from sklearn.metrics import recall_score
sensibilidad=recall_score(y_test,y_pred,average='binary',pos_label='f')
print('Sensibilidad del modelo')
print(sensibilidad)
```
